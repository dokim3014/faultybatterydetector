```python
import numpy as np
import matplotlib.pyplot as plt
import os
import scipy

from lib import tally, reorient
```

## Normal Import

```python
# Normal
path = r"C:\Users\user\Jupyter Works\20220526_응바공세미나팀플\XrayBat-master\gen_normal\output"
list_dir = os.listdir(path)

fname = []
for idx in range(len(list_dir)):
    if list_dir[idx].endswith('.22'):
        fname.append(list_dir[idx])

print(len(fname))
```

```python
dimension = (200, 200, 1)

vol_norm = np.zeros((len(fname), 40, 120), dtype=np.float32)
for idx in range(len(fname)):
    image_3d = tally.usrbin(os.path.join(path, fname[idx]), dimension)
    image_2d = image_3d[:,:,0]
    image_2d = (image_2d / image_2d.max() * -1) + 1
    img_clean = reorient.reorient(image_2d)
    img_norm = reorient.normalize(img_clean)
    vol_norm[idx] = img_norm
```

```python
plt.imshow(img_norm)
```

```python
#Displaying 465 images made the computer crash, so I split into 100 images
# No 0

starting_point = 0

size_test_single_vol = 100 #test_single_vol.shape[0]
print('size of validation dataset is %d' % size_test_single_vol)
fig, ax = plt.subplots(size_test_single_vol, 1, figsize=(5,size_test_single_vol*5))
for idx in range(size_test_single_vol):
    ax[idx].imshow(vol_norm[idx])
    ax[idx].set_title(fname[idx])
```
